```python
import pandas as pd 
import numpy as np 
import scipy.stats as st 
from scipy.interpolate import interp1d
from matplotlib import pyplot as plt 
import math
import statistics as stats

```

```python
data=pd.read_csv('sigma_U.csv',header=0,)
data = data.dropna(how='any')#quita los valores nan del documento
y=data.iloc[:,0]
z=data.iloc[:,1]
x=data.iloc[:,2]
w=data.iloc[:,3]


print(stats.stdev(z)) # desviación estandar 
#print(stats.stdev(w)) # desviación estandar 
print(data)
```

```python
yerror = z
xerror = w
plt.plot(x,y,'ro')
plt.errorbar(x,y,yerr=yerror, fmt = ' ' , ecolor='black')
plt.errorbar(x,y,xerr=xerror, fmt = ' ' , ecolor='blue')
plt.ylabel('sigma_U')
plt.xlabel('erg/seg')
plt.title("Luminosidad_UV_GALEX")
plt.xscale('log')
plt.yscale('log')
#plt.ylim(0,)
#plt.xlim(10e44,11e45)
plt.grid()
plt.savefig("Luminosidad_UV_GALEX_sigma_U_error.pdf")
plt.show
```
