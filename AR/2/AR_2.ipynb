```python
import pandas as pd 
import numpy as np 
import scipy.stats as st 
from scipy.interpolate import interp1d
from matplotlib import pyplot as plt 
import math
import statistics as stats
```

```python
data=pd.read_csv('AR_2_U.csv',header=0,)
data = data.dropna(how='any')#quita los valores nan del documento
y=data.iloc[:,0]
z=data.iloc[:,2]
x=data.iloc[:,1]
print(stats.stdev(y)) # desviación estandar 
print(data)
```

```python
yerror = 1.5939695052973406
xerror = z
plt.plot(x,y,'ro')
plt.errorbar(x,y,yerr=yerror, fmt = ' ' , ecolor='black')
plt.errorbar(x,y,xerr=xerror, fmt = ' ' , ecolor='blue')
plt.ylabel('AR_U_2')
plt.xlabel('erg/seg')
plt.title("Luminosidad_UV_GALEX_AB")
plt.xscale('log')
plt.grid()
plt.savefig("Luminosidad_UV_GALEX_AR_U_2_error.pdf")
plt.show
```
